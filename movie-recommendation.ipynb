```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
movies.head(2)
```

```python
movies.info()
```

```python
movies = movies[['title','overview','vote_count','vote_average']]
```

```python
movies
```

```python
num_votes = movies.groupby('title')['vote_count'].count().reset_index()
num_votes
```

```python
num_votes.rename(columns={'vote_count':'num_vote'},inplace=True)
num_votes
```

```python
avg_vote = movies.groupby('title')['vote_count'].mean().reset_index()
avg_vote.rename(columns={'vote_count':'avg_vote'},inplace=True)
avg_vote
```

```python
num_votes.merge(avg_vote,on='title').sort_values('avg_vote',ascending=False)
```

```python
popular_df = num_votes.merge(avg_vote,on='title')[['title','num_vote','avg_vote']]
popular_df
```

```python
popular_df = popular_df.sort_values('avg_vote',ascending=False).head(50)
```
